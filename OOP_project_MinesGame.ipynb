```python
class setting:
    width = 1227
    height = 720
    grid_size = 6
    cell_count = grid_size ** 2
    mines_count = (cell_count) / 4
```

```python
class utils:
    set = setting()
    def height_per(Prct):
         return (set.height / 100) * Prct
#--------------------------------------------------------------------------------- 
    def width_per(Prct):
         return (set.width / 100) * Prct
```

```python
from tkinter import Button
import random
import ctypes
import sys

class Cell:
    
    set =setting()
    
    all = []
    cell_count = set.cell_count
    cell_count_label_object = None
    def __init__(self, x, y, is_mine=False):
        self.is_mine = is_mine
        self.is_opened = False
        self.is_mine_cand = False
        self.cell_btn_object = None
        self.x = x
        self.y = y
        Cell.all.append(self)
#---------------------------------------------------------------------------------    
    def create_btn_object(self, location):
        btn = Button(
            location,
            width = 12,
            height = 4,
            
        )
        btn.bind('<Button-3>', self.right_click_actions)
        btn.bind('<Button-1>', self.left_click_actions)
        self.cell_btn_object =btn
#--------------------------------------------------------------------------------- 
    def right_click_actions(self,event):
        if not self.is_mine_cand:
            self.cell_btn_object.configure(bg='yellow')
            self.is_mine_cand = True
        else:
            self.cell_btn_object.configure(bg='SystemButtonFace')
            self.is_mine_cand = False
#--------------------------------------------------------------------------------- 
    
    def left_click_actions(self,event):
        if self.is_mine:
            self.show_mine()
        else:
            if self.surrounded_cells_mines_length == 0:
                for cell_obj in self.surrounded_cells:
                    cell_obj.show_cell()
            
            if self.cell_count == set.mines_count:
                ctypes.windll.user32.MessageBoxW(0,"Congratulations! You win","Game over",0)
        self.show_cell()
        
        #deni left and right click in the operned cells
        self.cell_btn_object.unbind('<Button-3>')
        self.cell_btn_object.unbind('<Button-1>')
            
#--------------------------------------------------------------------------------- 
    def show_mine(self):
        self.cell_btn_object.configure(bg='red')
        ctypes.windll.user32.MessageBoxW(0,"You clicked on a mine!","Game over",0)
        sys.exit()
        
#--------------------------------------------------------------------------------- 
    def get_cell_by_axis(self, x, y):
        #return value in form (X,Y)
        for cell in Cell.all:
            if cell.x == x and cell.y == y:
                return cell
#--------------------------------------------------------------------------------- 
    @property
    def surrounded_cells(self):
        cell = [
            self.get_cell_by_axis(self.x - 1, self.y - 1),
            self.get_cell_by_axis(self.x - 1, self.y),
            self.get_cell_by_axis(self.x - 1, self.y + 1),
            self.get_cell_by_axis(self.x, self.y - 1),
            self.get_cell_by_axis(self.x + 1, self.y - 1),
            self.get_cell_by_axis(self.x + 1, self.y),
            self.get_cell_by_axis(self.x + 1, self.y + 1),
            self.get_cell_by_axis(self.x, self.y + 1)
        ]
        cell = [ cell for cell in cell if cell is not None]
        return cell
 #---------------------------------------------------------------------------------    
    @property
    def surrounded_cells_mines_length(self):
        counter = 0
        for cell in self.surrounded_cells:
            if cell.is_mine:
                counter += 1
        return counter

#---------------------------------------------------------------------------------     
    def show_cell(self):
        if not self.is_opened:
            Cell.cell_count -= 1
            self.cell_btn_object.configure(text= self.surrounded_cells_mines_length)
    
            if Cell.cell_count_label_object:
                Cell.cell_count_label_object.configure(text= f"Cell Left: {Cell.cell_count}")
            
            self.cell_btn_object.configure(bg='SystemButtonFace')
        self.is_opened = True      

#---------------------------------------------------------------------------------    
    @staticmethod
    def create_cell_count_label(location):
        lbl = Label(
            location,
            bg='black',
            fg='white',
            text= f"Cell Left: {Cell.cell_count}",
            font= ("",27),
            width = 12,
            height= 4
        )      
        Cell.cell_count_label_object = lbl
#--------------------------------------------------------------------------------- 
    # to print it in a friendly way
    def __repr__(self):
        return f"Cell:({self.x},{self.y})"
#--------------------------------------------------------------------------------- 
    @staticmethod
    def randomize_mines():
        picked_cells = random.sample(Cell.all, int(set.mines_count))
        for picked_cell in picked_cells:
            picked_cell.is_mine = True
            
        
        
```

```python
from tkinter import *
import random
set = setting()
ut = utils()


root = Tk()
root.configure(bg='black')
root.geometry(f'{set.width}x{set.height}')
root.title("Minesweeper Game")
root.resizable(False,False)

#--------------------------------------------------------------------------------- 
top_frame = Frame(
    root,
    bg='black',
    width = set.width,
    height = utils.height_per(25)
)
top_frame.place(x=0, y=0)

#--------------------------------------------------------------------------------- 
left_frame = Frame(
    root,
    bg='black',
    width = utils.width_per(25),
    height = utils.height_per(75)
)
left_frame.place(x= utils.width_per(0), y= utils.height_per(25))
#--------------------------------------------------------------------------------- 
center_frame = Frame(
    root,
    bg='black',
    width = utils.width_per(75),
    height = utils.height_per(75)
)
center_frame.place(x= utils.width_per(25), y= utils.height_per(25))
#--------------------------------------------------------------------------------- 
game_title = Label(
    top_frame,
    bg='black',
    fg='white',
    text="MinesWeeper Game",
    font=47
)




for x in range(set.grid_size):
    for y in range(set.grid_size):
        c = Cell(x, y)
        c.create_btn_object(center_frame)
        c.cell_btn_object.grid(column= x, row= y)

Cell.create_cell_count_label(left_frame)
Cell.cell_count_label_object.place(x= 4, y= 3)





Cell.randomize_mines()





 


#to run the window
root.mainloop()
```
